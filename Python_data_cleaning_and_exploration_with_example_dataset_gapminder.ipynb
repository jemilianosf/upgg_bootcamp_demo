```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# histogram

plt.hist(gapminder_copy['lifexp'])
plt.xlabel('Life Expectancy')
plt.ylabel('Count')
```

```python
# boxplot
sns.boxplot(x = 'year', y = 'lifexp', data = gapminder_copy)
plt.xlabel('year')
plt.ylabel('life expectancy')
```
